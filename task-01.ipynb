```python
#Adjustment
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(40)

arr_size = 100000
exponential_data = np.random.exponential(size=arr_size)
uniform_data = np.random.rand(arr_size)
normal_data = np.random.randn(arr_size)

#Plotting 
plt.figure(figsize=(12,8))


#plotting histogram
bins = 100
plt.hist(uniform_data, bins=bins, density=True, histtype="step", color="blue", label="Uniform Data")
plt.hist(exponential_data, bins=bins, density=True, histtype="step", color="red", label="Exponential Data")
plt.hist(normal_data, bins=bins, density=True, histtype="step", color="green", label="Normal Data")

#Plot view adjustment
plt.axis([-4,6,0,1.1])
plt.legend(loc="upper left")
plt.title("Distribution comparison")
plt.xlabel("Index Values")
plt.ylabel("Density")

#show the plot 
plt.show()
```

### 3. Using Matplotlib, plot the Z = X2 + Y2 in a 3D plot. limit the x and y values to [-5,5]

```python
from mpl_toolkits.mplot3d import Axes3D

#Create data

#linspace offers evenly spaced numbers over a specified interval
#start, stop, num= number of samples to be created 
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y)
Z = X**2 - Y**2 #power operator 

#creating a 3D plot 
fig = plt.figure(figsize=(12,8)) 

#3D Axes 
ax = Axes3D(fig)

#plotting the 3D surface 
surf = ax.plot_surface(X,Y,Z, cmap=plt.cm.coolwarm, linewidth=0.1)
fig.colorbar(surf, shrink=0.5,aspect=5)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z = X^2 + Y^2')

plt.show()
```

### 4. Calculate the peason standard coefficient and spearman rank correlation for the [HP, Attack, Defense, Sp. Ark, Sp.Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make surethe values are displaced in the heatmap as well

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select the relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Calculate Pearson correlation
pearson_corr = df[features].corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = df[features].corr(method='spearman')

# Set up the matplotlib figure

# Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')
```

```python
# Create Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Rank Correlation Heatmap')

plt.show()
```
