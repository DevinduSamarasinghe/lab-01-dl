#### Download the Au_nanoParticle_dataset.csv and load the data of the csv file as a dataframe in a new colab notebook. Do the below subtask with the dataset. Use any of the above libraries if needed.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv('Au_nanoparticle_dataset.csv')
df.describe()
df.head()
```

#### 1. Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk, N_surface and R_avg columns.

```python
# Create a new dataframe with only the specified columns
feature_selected_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# Display the first few rows of the new dataframe
feature_selected_df.head()
```

#### 2. Display the first 20 samples of this dataframe.

```python
feature_selected_df.head(n=20)
```

#### 3. Calculate the mean, standard deviation and quartile values for each of the above 4 features.


```python
describe = feature_selected_df.describe().T

#mean 
print(describe['mean'],"\n")


#std
print(describe['std'],"\n")

#quartiles 
print(describe.iloc[:,-4:])
```

#### 4. Plot the histogram of each of these features in a 1x4 layout

```python
fig, axes = plt.subplots(1,4, figsize=(20,5))  

for i, col in enumerate(feature_selected_df.columns):
    axes[i].hist(feature_selected_df[col], bins=20, color='red')
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
    
plt.show()
```

#### 5. Visualize the scatter plots and histograms of this dataframe using the pairplot functionality of seaborn library.


```python
import seaborn as sns 
import pandas as pd 

sns.pairplot(feature_selected_df)
```

#### 6. Add the below code and change it such that, 

##### a. plots on the diagonal contains the histogram of each feature along with the kernel density estimation plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming filtered_df is your dataframe containing the specified columns
g = sns.PairGrid(feature_selected_df)

g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)

plt.show()
```
